### FFN

```python
from transformer import FFN, FFN_init
```

```python
ffn_init = FFN_init.randomInitial(
    d_model = d_model,
    expand_scale = 4
    )
```

```python
ffn1 = FFN(ffn_init)
```

```python
ffn1.W_expand
```

```python
forward = np.zeros((B, T, d_model))

rng = np.random.default_rng(40000)
backward = rng.random((B, T, d_model), dtype=np.float64)
```

```python
ffn1.forward_train(forward)
```

```python
ffn1.backward(backward)
```

```python
len(ffn1.d_W_expand_list)
```

```python
ffn1.save_checkpoint()
```

```python
copyffn1 = FFN.load_checkpoint("ffn1", activation_function=Relu, activation_function_backward=Relu_backward)
```

```python
copyffn1.forward_train(forward)
```
